```python
# Implement Gaussian Mixture Model (GMM) for expectation maximization
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.datasets import make_blobs

# Create a synthetic dataset with three clusters
X, y = make_blobs(n_samples=300, centers=3, random_state=42)

# Create a Gaussian Mixture Model with three components
gmm = GaussianMixture(n_components=3, random_state=42)

# Fit the model to the data using the Expectation-Maximization (EM) algorithm
gmm.fit(X)

# Predict the cluster assignments for each data point
y_pred = gmm.predict(X)

# Plot the original data and the GMM clusters
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis', marker='o', edgecolors='k', alpha=0.7)
plt.scatter(gmm.means_[:, 0], gmm.means_[:, 1], c='red', marker='x', s=100, linewidths=3, label='GMM Means')
plt.legend()
plt.title('Gaussian Mixture Model Clustering')
plt.show()
```
